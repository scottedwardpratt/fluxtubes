# Analytical Distribution Function:

Given:  
$$G_1 = \frac{1}{t}e^\frac{-(p-q)^2}{2a^2t}e^\frac{-(p+q)^2}{2b^2t}$$ 
and  
$$G_2 = \frac{1}{t}e^\frac{-(p-q)^2}{2b^2t}e^\frac{-(p_q)^2}{2a^2t}$$  
where:  
$$a^2 = \delta (p-q)^2$$  
and  
$$b^2 = \delta (p+q)^2$$  

Using the method of images we can determine our distribution function for a returning random walk $\rho (p,q,t)$ (where t is the number of steps taken) as:

$$ 
\rho = \frac{\partial^2 G_1}{\partial p \partial q} - \frac{\partial^2 G_2}{\partial p \partial q}
$$

This comes out to:

$$
\rho(p,q,t) = \frac{1}{t^3}[[\frac{(p+q)^2}{b^4}-\frac{(p-q)^2}{a^4}+t(\frac{1}{a^2}-\frac{1}{b^2})]e^\frac{-(p-q)^2}{2a^2t}e^\frac{-(p+q)^2}{2b^2t}+[\frac{(p+q)^2}{a^4}-\frac{(p-q)^2}{b^4}+t(\frac{1}{b^2}-\frac{1}{a^2})]e^\frac{-(p-q)^2}{2b^2t}e^\frac{-(p+q)^2}{2a^2t}]
$$

Which is frankly a monstrosity. If we introduce "nicer" coordinates $\lambda = (p+q)$ and $\mu = (p-q)$, we get the slightly more manageable form of:

$$
\rho(\lambda,\mu,t) = \frac{1}{t^3}[[\frac{\lambda^2}{b^4}-\frac{\mu^2}{a^4}+t(\frac{1}{a^2}-\frac{1}{b^2})]e^\frac{-\mu^2}{2a^2t}e^\frac{-\lambda^2}{2b^2t}+[\frac{\lambda^2}{a^4}-\frac{\mu^2}{b^4}+t(\frac{1}{b^2}-\frac{1}{a^2})]e^\frac{-\mu^2}{2b^2t}e^\frac{-\lambda^2}{2a^2t}]
$$

Some validation that this is correct is to consider the case where $a=b=D$

$$
\rho(p,q,t) = \frac{1}{t^3}[[\frac{(p+q)^2}{D^4}-\frac{(p-q)^2}{D^4}+t(\frac{1}{D^2}-\frac{1}{D^2})]e^\frac{-(p-q)^2}{2D^2t}e^\frac{-(p+q)^2}{2D^2t}+[\frac{(p+q)^2}{D^4}-\frac{(p-q)^2}{D^4}+t(\frac{1}{D^2}-\frac{1}{D^2})]e^\frac{-(p-q)^2}{2D^2t}e^\frac{-(p+q)^2}{2D^2t}]
$$

$$
\rho(p,q,t) = \frac{1}{t^3}[[\frac{(p+q)^2}{D^4}-\frac{(p-q)^2}{D^4}]+[\frac{(p+q)^2}{D^4}-\frac{(p-q)^2}{D^4}]]e^\frac{-(p+q)^2}{2D^2t}e^\frac{-(p-q)^2}{2D^2t}
$$

$$
\rho(p,q,t) = \frac{8pq}{t^3D^4}e^\frac{-p^2}{D^2t}e^\frac{-q^2}{2D^2t}
$$

This is a simple product of gaussians linear in p and q. I expect this to be what the distribution function would look like for a completely independent returning random walk $(a=b)$

One can also check that $\rho(p=0,q,t) = 0$:

$$
\rho(p=0,q,t) = \frac{1}{t^3}[[\frac{q^2}{b^4}-\frac{q^2}{a^4}+t(\frac{1}{a^2}-\frac{1}{b^2})]e^\frac{-q^2}{2a^2t}e^\frac{-q^2}{2b^2t}+[\frac{q^2}{a^4}-\frac{q^2}{b^4}+t(\frac{1}{b^2}-\frac{1}{a^2})]e^\frac{-q^2}{2b^2t}e^\frac{-q^2}{2a^2t}]
$$

$$
\rho(p=0,q,t) = \frac{1}{t^3}[[\frac{q^2}{b^4}-\frac{q^2}{a^4}+t(\frac{1}{a^2}-\frac{1}{b^2})]+[\frac{q^2}{a^4}-\frac{q^2}{b^4}+t(\frac{1}{b^2}-\frac{1}{a^2})]]e^\frac{-q^2}{2b^2t}e^\frac{-q^2}{2a^2t}
$$

$$
\rho(p=0,q,t) = \frac{1}{t^3}[[\frac{q^2}{b^4}-\frac{q^2}{a^4}+t(\frac{1}{a^2}-\frac{1}{b^2})]-[\frac{q^2}{b^4}-\frac{q^2}{a^4}+t(\frac{1}{a^2}-\frac{1}{b^2})]]e^\frac{-q^2}{2b^2t}e^\frac{-q^2}{2a^2t}
$$

$$
\rho(p=0,q,t) = 0
$$

A similar check can be performed for $\rho(p,q=0,t) = 0$.

Thus, the analytic form of:

$$
\rho(\lambda,\mu,t) = \frac{1}{t^3}[[\frac{\lambda^2}{b^4}-\frac{\mu^2}{a^4}+t(\frac{1}{a^2}-\frac{1}{b^2})]e^\frac{-\mu^2}{2a^2t}e^\frac{-\lambda^2}{2b^2t}+[\frac{\lambda^2}{a^4}-\frac{\mu^2}{b^4}+t(\frac{1}{b^2}-\frac{1}{a^2})]e^\frac{-\mu^2}{2b^2t}e^\frac{-\lambda^2}{2a^2t}]
$$

Seems to fit the bill.
